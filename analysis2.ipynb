```python
df = df.drop(["Forest area", "Individuals using the Internet", "People practicing open defecation", "Electric power consumption"], axis = 1)
df
```

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df,
    x="Life Expectancy", y="People using at least basic drinking water services",
    hue="Life Expectancy", size="People using at least basic drinking water services",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
sns.barplot(x='Life Expectancy', y='Obesity among adults',data=df)
plt.show()
```
