It appears as if the dataset is perfect with no missing values at all. However, when we analyse the dataset we can see that there are few data entries which are zero but don't make sense to be zero. For example, in many rows in the variable open defecation we can observe zeros and few zero is military expenditure as well. However, we understand that these two value are very unlikely to be zero ever hence, the author must have just put zero in place of missing values. Hence, we have replaced zero values with np.nan suggesting missing values. 
Additionally, I noticed that there are a few zero values in the column Beer consumption per capita. Upon further analysis I found the zeros were in Islamic countries like Bangladesh and Iran, it is indeed banned in those countries and it makes sense for zero values to be in those data entries. Hence, I decided to not replace zero values in the column Beer consumption per capita.

```python
cols_to_replace = [col for col in df.columns if col != 'Beer consumption per capita']

df[cols_to_replace] = df[cols_to_replace].replace(0, np.nan)
```

```python
msno.bar(df)
```

```python
msno.matrix(df)
```

Once we have replaced zero values with missing values in all the columns except beer consumption per capita we can indeed there are a few missing values in the dataset, especially in the column people practicing open defecation.

Let's make a correlation matrix for the dataframe


```python
data = df.corr()
plt.figure(figsize=(20, 15))
plt.title('Correlation Matrix of the datframe')
sns.heatmap(data, annot=True, cmap='plasma')
```

As our target variable is CO2 emission, we can see that the few columns' correlation coefficient with CO2 emissions is pretty weak. Columns Year, Popoulation and Millitary Expenditure have very weak correlation coeficcienct, -0.0078, -0.052 and 0.046, respectively. Hence, I have decided to drop them from the analysis. Fuurthermore, as the column People practicing open defecation has many missing values I have decided to drop that column as well.


```python
df_cl = df.drop(['Year', 'Population', 'Military expenditure', 'People practicing open defecation'], axis=1)
```

```python
df_cl= df_cl.dropna()
df_cl
```

Now Let's see the spread of the data for each of the columns using a violinplot.
Violinplot has been chosen over boxplot because violinplot not only depicts the statistical summary but also the density of each variable.

```python
# Loop over each numeric column and create a boxplot
#for col in numeric_cols:
 #   plt.boxplot(df_cl[col])
  #  plt.title(col)
   # plt.show()
    
numeric_cols = df_cl.select_dtypes(include='number').columns.tolist()

fig, axs = plt.subplots(nrows=int(np.ceil(len(numeric_cols)/2)), ncols=2, figsize=(15, 20))
#Printing in a table like format with 2 columns
for i, col in enumerate(numeric_cols):
    ax = axs[i//2, i%2]
    ax.violinplot(df_cl[col])
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

Let's look at the pairplot graph in order to view the relation between the variables
