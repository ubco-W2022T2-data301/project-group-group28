## Method Chaining

Now that we understand what steps are required to clean the data, let's do this entire process of cleaning the data in one step.

```python
def load_data(raw_data_path):
    return pd.read_csv(raw_data_path, sep=';')

def load_and_wrangle_data(raw_data_path):
    """
    This function loads the data from the specified raw_data_path and does some wrangling by dropping unnecessary columns, 
    creating a binary column for high GDP per capita, replacing 0 values with NaNs in non-beer consumption columns, 
    dropping rows with NaNs and renaming the high_gdp column.
    """
    df = (load_data(raw_data_path)
          .drop(['Year', 'Population', 'Military expenditure', 'People practicing open defecation'], axis=1)
          .assign(high_gdp=lambda x: x['GDP per capita'] > 12235)
          .apply(lambda x: x.replace(0, np.nan) if x.name != 'Beer consumption per capita' else x, axis=0)
          .dropna()
          .rename(columns={'high_gdp': 'High GDP per cap'})
         )
    return df
```

Here I have created two function, first function to load the raw data and the second function to process the loaded to data. Here I call the load_data function in load_and_wrangle_data function so that entire process of loading and wrangling data can be peformed simply by calling the load_and_wrangle_data as done below.

```python
import Project_Functions1 as pf


df = pf.load_and_wrangle_data("../data/raw/Life_Expectancy_00_15.csv")
df
```

### Target variable CO<sub>2</sub> Emissions Analysis

Let's start the exploration by exploring the spread of the target variable i.e., carbon dioxide emissions.

```python
plt.hist(df["CO2 emissions"], bins=30, color="blue")

# add labels and title
plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.title("Histogram of CO2 Emissions")

# show the plot
plt.show()
```

In the above histogram we can see that the spread is positively skewed. This spread suggests that for most of the entries the carbon dioxide emissions is low, and there are only a few entries with high carbon dioxide emissions.

As we are analysing factors influencing Carbon dioxide emissions, let's explore if any particular continent stands out in terms of carbon dioxide emissions.

```python
co2_by_continent = df.loc[:, ['Continent', 'CO2 emissions']]
co2_by_continent.boxplot(by='Continent', figsize=(10, 6)) #groupby continent
plt.title('Box plot of CO2 emissions by continent')
plt.xlabel('Continent')
plt.ylabel('CO2 emissions')
plt.show()
```

In the above boxplot, we have grouped the entries by continent, and we can observe that there are certainly some continents that have the higher carbon dioxide emissions than other. On one hand we can observe that on an average the continent of Oceania is one with highest carbon dioxide emissions, on the other hand, we can observe that the continent of Africa has the lowest carbon dioxide emissions.

Saying so, we must also note that what we discussed is averages, but continents like Asia display some anomalous behaviour as they have outliers. Even though, on an average, Asia is not the continent with the highest carbon dioxide emissions, one of the countries in Asia has the highest carbon dioxide emissions in the world. 


Below we explore if any there is any relation between a country being developed and it's carbon dioxide emissions.

```python
plt.title('Box plot of CO2 emissions by development status')
sns.boxplot(x='Least Developed', y='CO2 emissions', data=df)
plt.figure(figsize=(10,20)) 
plt.show()
```

In the above boxplot we can see that there is a certainly a correlation between the development status of a country and carbon dioxide emissions.

Through the graph above we can confidently say that the countries that are developed have higher carbond dioxide emissions that countries that are not developed.

Here we explore the relation between carbon dioxide emissions, electric power consumption and GDP per capita.

```python
sns.scatterplot(data=df, y='CO2 emissions', x='Electric power consumption', hue='High GDP per capita', alpha=0.3) 
#setting alpha to resolve overplotting issue
plt.title('Relationship between CO2 emissions, Electric power consumption and GDP per capita')
plt.show()
```

We were bound find a corelation between electric power consumption per and carbon dioxide emissions, as electric power consumption increases, carbon dioxide emissions. Moreover, the new insight we can observe here is that the economic development of country which is indicated through GDP per capita also has an influence over carbon dioxide emissions.

Let's explore this new insigh in depth below:

```python
sns.scatterplot(data=df, y='CO2 emissions', x='GDP per capita', alpha=0.3)
plt.title('Relationship between CO2 emissions and GDP per capita')
plt.show()
```

In the scatter plot above, we can see a strong correlation between GDP per capita and carbon dioxide emissions. As GDP per capita increases, the carbon dioxide emissions also increase, both the variables move in the same direction.

One can argue that the relation between these two variables is close to a linear relaiton. Let's explore this relation further through a regression plot below:

```python
sns.regplot(data = df, x='GDP per capita', y='CO2 emissions', scatter_kws={'alpha':0.3})

plt.xlabel('GDP per capita')
plt.ylabel('CO2 emissions')
plt.title('Regression plot between CO2 emissions and GDP per capita')

plt.show()
```

```python
corr = df['GDP per capita'].corr(df['CO2 emissions'])

print(corr)
```

Through the regression plot we can see that the relaiton between the two variable is close to a linear relation. Looking at the correlation matrix plotted early on the analysis or as as printed above, is 0.828 which suggests a strong positive correlation between the two variables GDP per capita and carbon dioxide emissions.
