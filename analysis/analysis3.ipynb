```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("../data/raw/Life_Expectancy_00_15.csv", sep=';')
gdp_health_df = df[['GDP per capita', 'Health expenditure']]
sns.set_style('whitegrid')
g = sns.jointplot(data=gdp_health_df, x='GDP per capita', y='Health expenditure', kind='hex', cmap='viridis', gridsize=15)
g.fig.set_size_inches(8, 6)  
plt.xlabel('GDP per capita')
plt.ylabel('Health Expenditure')
plt.title('Hex Bin Plot of GDP vs Health Expenditure', y=1.2)  
plt.show()
```

Surprisingly,we can observe from the above plot, that countries with really high GDP per capita have the lowest Health Expenditure,eventhough one might be lead to believe they would have the highest Health Expenditure due to their immense wealth.But rather surprisingly countires with low middle GDP per capita of around 30000-50000 have the highest Health Expenditure.This might be due to the super rich countries like the USA and Qatar spending money overwhelming on other sectors like (military for USA) and (petroleum extraction for Qatar) while socialist nordic countries who pride themselves on their healthcare obviously spend a lot more on their Health sector.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("../data/raw/Life_Expectancy_00_15.csv", sep=';')
high_gdp_df = df[df['GDP per capita'] > np.median(df['GDP per capita'])][['GDP per capita', 'Health expenditure', 'CO2 emissions', 'Forest area', 'Electric power consumption', 'Individuals using the Internet']]
corr_matrix = high_gdp_df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(data=corr_matrix, annot=True, mask=mask, cmap='coolwarm', square=True)
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.title('Heatmap of Correlation Matrix for High GDP Countries')
plt.show()
```

As expected countries with the highest GDP per capita have greater CO2 emissions as both the variables are overwhelmingly positively correlated(+0.78 correlation).Another expected feature is the GDP per capita and Electric power consumption being strongly positively correlated(+0.67 correlation) since a richer a country is,the more electricity it will utilise for it's industries and housing for it's citizens.But the most surprising aspect of the Heatmap is that GDP per capita and Health Expenditure have very weak correlation,which as analysed in the above scatterplot is due to the richest countries different priorities in spending on different sectors.Another interesting feature I noticed is that GDP per capita and Forest area are also weakly correlated, which makes sense as some of the richest countries are Middle Eastern countries,that have no fertile land eg: Qatar,Saudia Arabia and Oman.Lastly, I also noticed that GDP per capita is weak to moderately positively correlated with Individuals using the internet, as obviously a richer a country the higher quality of life thus more access to internet but it's comparatively weak due to rich countries like China's strict censorship policies.

```python
#Wrapping my method chains in a function
def load_and_process(file_path):

    # Method Chain 1 (Load data and remove unused columns)

    df1 = (
          pd.read_csv(file_path, sep=';')
          .drop(['Continent', 'Least Developed', 'Population', 'People practicing open defecation', 
                 'People using at least basic drinking water services', 'Obesity among adults', 
                 'Beer consumption per capita'], axis=1)
          .dropna()
      )

    # Method Chain 2 (Create new column expenditure_type)

    df2 = (
          df1
          .assign(expenditure_type=lambda x: np.where(x['Health expenditure'] > x['Military expenditure'], 'H', 'M'))
      )

    return df2


```

As we can view from below countries with the highest GDP per capita are the ones with least amount of health expenditure,infact on par with countries with the lowest GDP per capita.The reason is fairly straightforward since countries with such high GDP per capita like the USA and Qatar spending money overwhelming on other sectors like (military for USA) and (petroleum extraction for Qatar) while socialist nordic countries who pride themselves on their healthcare obviously spend a lot more on their Health sector. And since these countries are active on the geopoliical scale and engage in proxy warfare they spend a whopping amount on their military as can been seen below, where many times Qatar's military spending trumped their health spending.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("../data/raw/Life_Expectancy_00_15.csv", sep=';')
gdp_health_df = df[['GDP per capita', 'Health expenditure']]
sns.jointplot(data=gdp_health_df, x='GDP per capita', y='Health expenditure', kind='hex', cmap='viridis', gridsize=15)
plt.xlabel('GDP per capita')
plt.ylabel('Health Expenditure')
plt.title('Hex Bin Plot of GDP vs Health Expenditure')
plt.title('Hex Bin Plot of GDP vs Health Expenditure', y=1.2) 
plt.show()
```

```python
import matplotlib.pyplot as plt

usa = df.loc[df['Country'] == 'Qatar', ['Year', 'Military expenditure', 'Health expenditure']]
usa = usa.set_index('Year')

ax = usa.plot(kind='bar', figsize=(10, 6))
ax.set_ylabel('Expenditure (% of GDP)')
ax.set_xlabel('Year')
ax.set_title('Military and Health Expenditure of Qatar')
plt.show()
```

As we can see from the line graph below countries with the highest GDP per capita like USA,Qatar,Singapore and Hong Kong have the highest CO2 emissions,which was expected since they are highly industrialised thus more factories emitting greenhouse gases like CO2.In fact intially CO2 emissions are directly proportional to GDP per capita and highly positively correlated. However a few countries with GDP per capita have relatively low CO2 emissions like Norway and Canada,since they take special measures to reduce global warming as they are at most risk of being affected by global warming due to them being in the Arctic.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = df.dropna(subset=['CO2 emissions', 'GDP per capita'])


fig, ax = plt.subplots(figsize=(10, 6))
hb = ax.hexbin(df['GDP per capita'], df['CO2 emissions'], gridsize=15, cmap='viridis', mincnt=1)
ax.set_xlabel('GDP per capita (current US$)')
ax.set_ylabel('CO2 emissions (metric tons per capita)')
ax.set_title('CO2 emissions vs GDP per capita (Hex Bin Plot)')
cb = plt.colorbar(hb)
cb.set_label('Count')
plt.show()

```

We can infer from the below graph that super rich countries(high GDP per capita countries) have extremely low forrest area since many of these countries are micro nations like Singapore,Malta,Luxembourg,Hong Kong and Qatar.Thus they can't afford to dedicate a sizable portion of their land for nature or greenery and thus comprimise it for industries like Petroleum and Oil rigs.While really low GDP countries also have low forrest area since most of these countries are African countries,and since Africa as a whole isn't fertile thus their forest area is low.While mid-level GDP countries like Vietnam and Brazil have really high forrest area since their geogrpahy is highly tropical thus perfect for forestry.

Rich countries(high GDP countries) have a strong postive correlation with CO2 emissions and Electric power consumption and light positive correlation with indiviuals using the internet.I already explained the reason for CO2 emissions, while Electric power consumption and indiviuals using the internet are strong signs of a developed country as they indicate a good quality of life as greater number of people have access to resources like internet and television.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("../data/raw/Life_Expectancy_00_15.csv", sep=';')
high_gdp_df = df[df['GDP per capita'] > np.median(df['GDP per capita'])][['GDP per capita', 'Health expenditure', 'CO2 emissions', 'Forest area', 'Electric power consumption', 'Individuals using the Internet']]
corr_matrix = high_gdp_df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  
sns.heatmap(data=corr_matrix, annot=True, mask=mask, cmap='coolwarm', square=True)
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.title('Heatmap of Correlation Matrix for High GDP Countries')
plt.show()
```
